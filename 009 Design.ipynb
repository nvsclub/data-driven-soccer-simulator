```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, FancyArrowPatch
from matplotlib.path import Path
```

```python
plt.rcParams['figure.figsize'] = (14,4)
plt.rcParams['figure.dpi'] = 1000
```

```python
sq_linewidth = 2.5
rect_size = 14
h_rs = rect_size/2
margin = 0.5
init_loc = 10

plt.xticks([])
plt.yticks([])
plt.box(False)
plt.xlim([-5, 105])
plt.ylim([40, 92])

patches = []
patches.append(Rectangle((50-h_rs,80), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(50, 85, 'State\n(x,y)', va='center', ha='center')
patches.append(FancyArrowPatch((50-h_rs,80), (init_loc+rect_size,70), arrowstyle='fancy', color='Black', mutation_scale=15))
patches.append(FancyArrowPatch((50,80), (50,70), arrowstyle='fancy', color='Black', mutation_scale=15))
patches.append(FancyArrowPatch((50+h_rs,80), (100-init_loc-rect_size,70), arrowstyle='fancy', color='Black', mutation_scale=15))

patches.append(Rectangle((init_loc,60), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(init_loc+h_rs, 65, 'Pass\n(endX,endY)', va='center', ha='center')
patches.append(FancyArrowPatch((init_loc,60), (init_loc-margin,50), arrowstyle='fancy', color='Black', mutation_scale=15))
plt.text(init_loc, 55, '\nSuccess', va='center')
patches.append(FancyArrowPatch((init_loc+rect_size,60), (init_loc+rect_size+margin,50), arrowstyle='fancy', color='Black', mutation_scale=15))
plt.text(init_loc+rect_size+margin, 55, '\nUnsuccess', va='center')
patches.append(Rectangle((50-h_rs,60), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(50, 65, 'Dribble\n(endX,endY)', va='center', ha='center')
patches.append(FancyArrowPatch((50-h_rs,60), (50-h_rs-margin,50), arrowstyle='fancy', color='Black', mutation_scale=15))
plt.text(50-h_rs, 55, '\nSuccess', va='center')
patches.append(FancyArrowPatch((50+h_rs,60), (50+margin+h_rs,50), arrowstyle='fancy', color='Black', mutation_scale=15))
plt.text(50+h_rs+margin, 55, '\nUnsuccess', va='center')
patches.append(Rectangle((100-init_loc-rect_size,60), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(100-init_loc-h_rs, 65, 'Shot', va='center', ha='center')
patches.append(FancyArrowPatch((100-init_loc-h_rs,60), (100-init_loc-h_rs,50), arrowstyle='fancy', color='Black', mutation_scale=15))

patches.append(Rectangle((init_loc-h_rs-margin,40), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(init_loc-margin, 45, 'State\n(endX,endY)', va='center', ha='center')
patches.append(Rectangle((init_loc+h_rs+margin,40), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(init_loc+rect_size+margin, 45, 'End', va='center', ha='center')
patches.append(Rectangle((50-rect_size-margin,40), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(50-h_rs-margin, 45, 'State\n(endX,endY)', va='center', ha='center')
patches.append(Rectangle((50+margin,40), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(50+margin+h_rs, 45, 'End', va='center', ha='center')
patches.append(Rectangle((100-init_loc-rect_size,40), rect_size, 10, linewidth = sq_linewidth, fill=False, color='Black'))
plt.text(100-init_loc-h_rs, 45, 'Reward', va='center', ha='center')

for patch in patches:
    plt.gcf().gca().add_artist(patch)

plt.savefig('mams report/diagram.eps', format = 'eps')
plt.savefig('mams report/diagram.png', format = 'png', dpi=500)
```
